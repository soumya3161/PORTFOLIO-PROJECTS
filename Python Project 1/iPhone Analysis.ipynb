```python
import pandas as pd
```

```python
df=pd.read_csv("iphone.csv")
df
```

### The column names have spaces.Renaming the column names to have underscore '_' instead of space

```python
df.columns=df.columns.str.replace(' ','_')
df
```

### Star rating for some of the models is missing in the dataset. filling those missing values with the average rating all the models.

```python
avg=df['Star_Rating'].mean(skipna=True)
df['Star_Rating'].fillna(value=avg)
```

### Now instead of filling missing values with avg rating of full dataset , fill with avg rating based on RAM. example : if rating for a 2 gb phone is missing then taking the average of all other 2 gb phones rating and fill that value.

```python
df_ram_avg=df.groupby('Ram',as_index=False).agg({'Star_Rating':'mean'})
df_merged=pd.merge(df,df_ram_avg,how='inner',on='Ram')
df_merged['Star_Rating_x'].fillna(value=df_merged['Star_Rating_y'],inplace=True)
df_merged.drop(columns='Star_Rating_y',inplace=True)
df_merged.columns=df.columns
df_merged['Star_Rating']=df_merged['Star_Rating'].round(2)
df_merged
```

### Creating a new column in the dataframe "Discount_Percentage" based on MRP and sale value

```python
df['Discount_Percentage']=(((df['Mrp']-df['Sale_Price'])/df['Mrp'])*100).round(2)
df
```

### Model has highest percent discount ?

```python
df.loc[df['Discount_Percentage']==df['Discount_Percentage'].max(),['Product_Name','Discount_Percentage']]
```

### Total no of models each space configuration (128 GB , 64 GB etc)

```python
def extract_space_configuration(Product_Name):
    parts=Product_Name.split('(')
    if len(parts)>1:
        part=parts[1].strip()
        if ',' in part:
            space_part=part.split(',')
            space=space_part[1].strip()
            space_conf=space[:len(space)-1]
            return space_conf
    return None
df['Space_Configuration']=df['Product_Name'].apply(extract_space_configuration)
df_space_conf=df['Space_Configuration'].value_counts().reset_index()
df_space_conf       
```

7- find total number of models for each color

```python
def extract_color(Product_Name):
    parts=Product_Name.split('(')
    if len(parts)>1:
        part=parts[1].strip()
        if ',' in part:
            color_part=part.split(',')
            color=color_part[0].strip()
            return color
    return None
df['iPhone_color']=df['Product_Name'].apply(extract_color)
df_color_count=df['iPhone_color'].value_counts().reset_index()
df_color_count
```

### Total number of models by iphone version : eg
#### iPhone 8: 9
#### iPhone XR : 5
#### so on..

```python
def extract_version(Product_Name):
    parts=Product_Name.split('iPhone')
    if len(parts)>1:
        part=parts[1].strip()
        if ' ' in part:
            version_part=part.split(' ')
            version=version_part[0].strip()
            return 'iPhone '+version
    return None
df['Version']=df['Product_Name'].apply(extract_version)
df_version_count=df['Version'].value_counts().reset_index()       
df_version_count
```

### Total no of reviews for iphone 11 and iphone 12 category . Output should have only 2 rows (for 11 and 12).

```python
# 1st approach
df['category']='other'
df.loc[df['Product_Name'].str.contains('iPhone 11'),'category']='iPhone 11'
df.loc[df['Product_Name'].str.contains('iPhone 12'),'category']='iPhone 12'
df_category=df[df['category'].isin(['iPhone 11','iPhone 12'])]
df_category.groupby('category',as_index=False)['Number_Of_Reviews'].sum()

# 2nd approach
iPhone_11=df.loc[df['Product_Name'].str.contains('iPhone 11'),'Number_Of_Reviews'].sum()
iPhone_12=df.loc[df['Product_Name'].str.contains('iPhone 12'),'Number_Of_Reviews'].sum()
df_total_no_of_reviews=pd.DataFrame({'Category':['iPhone 11','iPhone 12'],'Number_Of_Reviews':[iPhone_11,iPhone_12]})
df_total_no_of_reviews
```
